# Fifa Data **Analysis**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv("players_20.csv")
```

```python
fifa.head()
```

```python
fifa.shape
```

```python
for col in fifa.columns:
  print(col)
```

```python
fifa['nationality_name'].value_counts()[0:10]
```

```python
# no. of players from different nationality
plt.figure(figsize=(5,5))
sns.barplot(x=fifa['nationality_name'].value_counts()[0:5].keys() , y=fifa['nationality_name'].value_counts()[0:5], palette='magma')
plt.xlabel('nationality_name')
plt.show()
```

```python
player_sal = fifa[['short_name', 'wage_eur']]
```

```python
player_sal.head()
```

```python
player_sal = player_sal.sort_values(by=['wage_eur'], ascending=False)
```

```python
player_sal.head()
```

```python
# plotting wages of player according of descending wages
plt.figure(figsize=(7,5))
sns.barplot(x=fifa['short_name'][0:5] , y=fifa['wage_eur'][0:5] , palette='magma')
plt.grid(True)
plt.show()
```
